```r
library(phangorn)
```

## Neighbor-joining tree on variable sites only

```r
aln <- read.phyDat("../output/merged_var_subset_ontarget.fa", format="fasta")
```

```r
tree <- NJ(dist.ml(aln))
```

```r
rooted_tree <- root(tree, outgroup="Chimp", resolve.root=TRUE)
```

```r
plot(rooted_tree)
```

```r
plot(rooted_tree, use.edge.length=FALSE)
```
