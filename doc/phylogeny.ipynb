```r
library(phangorn)
```

```r
targets <- "lippold"
```

## B-team

```r
aln <- read.phyDat(paste0("../fasta/", targets, "_bteam.fa"), format="fasta")
```

```r
tree <- NJ(dist.ml(aln))
```

```r
rooted_tree <- root(tree, outgroup=c("A00_1", "A00_2"), resolve.root=TRUE)
```

```r
plot(rooted_tree, use.edge.length=FALSE)
```

```r
plot(rooted_tree, use.edge.length=FALSE)
```
