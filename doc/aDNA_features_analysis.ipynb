```python
def plot_damage(bam_file, ylim=30):
    fwd_mismatches, rev_mismatches = analyze_bam(bam_file)
    _ = plot_mismatches(fwd_mismatches, 'damage on forward -- ' + bam_file, ylim)
    _ = plot_mismatches(rev_mismatches, 'damage on reversed -- ' + bam_file, ylim)
```
